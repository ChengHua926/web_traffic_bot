```python
import random
thisString = "January"
traffic_data = {}

for i in range(1,32):
    number = random.randint(80000, 100000)
    number2 = random.randint(70000, 120000)
    index = random.randint(1, 100)
    if index % 2 == 0:
        dictNumber = number
    else:
        dictNumber = number2
    traffic_data[thisString + " " + str(i)] = dictNumber 
```

```python
import matplotlib.pyplot as plt


dates = list(traffic_data.keys())
traffic_counts = list(traffic_data.values())

# set up the figure size and title
fig = plt.figure(figsize=(8, 4))
fig.suptitle('Website Traffic in January 2023', fontsize=14, fontweight='bold')

# create a line graph of website traffic over the month
plt.plot(dates, traffic_counts, color='blue', linewidth=2, marker='o', markersize=6, markerfacecolor='white', markeredgecolor='blue')

# add grid lines and set axis limits
plt.grid(True)
plt.ylim(0, 130000)

# set the x-axis labels and rotate them for better visibility
plt.xticks(rotation=90, fontsize=8)

# set the y-axis label and tick format
plt.ylabel('Traffic Count', fontsize=12)
plt.ticklabel_format(style='plain', axis='y', useOffset=False)

# add data labels to the markers
for i, count in enumerate(traffic_counts):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=8)

# show the graph
plt.tight_layout()
plt.show()
```
